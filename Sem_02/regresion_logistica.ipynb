# In Synthetic data

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_classification
```

```python
X, y = make_classification(n_samples = 500  , n_features = 2 , n_redundant = 0,
                           n_informative = 2, n_clusters_per_class = 1, n_classes = 2, random_state = 8)
print(X.shape, y.shape)
_, ax = plt.subplots(figsize=( 5,4 ))
# changed ax.scatter( X , y , c = "r" , cmap = 'Paired')
# to use the two columns of X as x, y and use the target variable y as c argument.
ax.scatter( X[:,0] , X[:,1] , c = y , edgecolors="k", cmap = 'Dark2_r')
```

```python
from math import log
# Create an instance
logistic_regression = LogisticRegression()

# Train the model
logistic_regression.fit ( X , y ) #(# Samples, # Features)
```

```python
print(logistic_regression.coef_)
print(logistic_regression.intercept_)
```

```python
# Min-Max values for the grid of decision boundaries
offset=0.5
h_min, h_max = min(X[:,0])-offset , max(X[:,0])+offset
v_min, v_max = min(X[:,1])-offset , max(X[:,1])+offset
# Create grid
h_grid, v_grid = np.meshgrid( np.linspace(h_min,h_max,100),np.linspace(v_min,v_max,100) )
print(h_grid.shape, v_grid.shape)
print(h_grid.ravel().shape, v_grid.ravel().shape)
print(np.concatenate((h_grid.ravel().reshape(-1,1),v_grid.ravel().reshape(-1,1)),axis=1).shape)
print(np.c_[h_grid.ravel(),v_grid.ravel()].shape)

# Estimate predictions over the grid
pre_grid = logistic_regression.predict(np.c_[h_grid.ravel(), v_grid.ravel()])
print (pre_grid.shape)
pre_grid = pre_grid.reshape(h_grid.shape)
print (pre_grid.shape)
```

```python
# Draw grid
_, ax = plt.subplots (figsize=(6,5))
ax.pcolormesh( h_grid , v_grid , pre_grid , cmap = "Dark2_r")

# Draw data

# Scatter real values
ax.scatter ( X[:,0] ,X[:,1], c = y, edgecolors = 'k', cmap = "Dark2_r")

ax.set_xticks(())
ax.set_yticks(())
```
