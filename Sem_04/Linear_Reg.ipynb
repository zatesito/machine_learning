# **Linear Regresion**

```linear_regression``` module documentation [here](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html?highlight=linear%20regression#sklearn.linear_model.LinearRegression)

````python
import numpy as np # Numerical library
import matplotlib.pyplot as plt # plots
from sklearn.linear_model import LinearRegression
````

````python
def f(x) : # Create a function that samples from linear function
  y = (0.0918 * x) + 1.2859 + 0.1*np.random.randn (x.shape[0]) # Linear function
  return y

# create training samples
X = np.arange (0, 20, 0.25); print (X.shape)

y = f(X); print (y.shape)


# Scatter plot
plt.scatter(X, y, color = 'blue')
````

````python
# Create an instance
linear_regression = LinearRegression()

# Train the model
linear_regression.fit(X.reshape(-1,1), y) # (# Samples, # Features)

# Print estimated values
print ("w = " + str(linear_regression.coef_) + ", b = " + str(linear_regression.intercept_))
````

````python
# Create a new sample estimate prediction
new_sample = np.array([5])
print (f(new_sample))
print(new_sample.shape)

prediction = linear_regression.predict( new_sample.reshape(-1,1) )
print (prediction)
````

````python
# Predict x values
predictions = linear_regression.predict( X.reshape(-1,1) )

# Draw real samples (blue) and predictions (red)
_, ax = plt.subplots (figsize = ( 5 , 4 ))
ax.scatter( X , y, color = 'blue')
ax.plot ( X , predictions, color = 'red', linestyle = "--")
````

## In real data

```datasets``` module info [here](https://scikit-learn.org/stable/modules/classes.html#module-sklearn.datasets)

````python
import numpy as np
import matplotlib.pyplot as plt # plots
from sklearn.linear_model import LinearRegression # scikit-learn: libreria de IA para python
from sklearn.datasets import load_diabetes
````

````python
# Cargar la base de datos Diabetes
diabetes_X, diabetes_y = load_diabetes(return_X_y=True)
print(diabetes_X.shape, diabetes_y.shape)
````

````python
# Use only one feature
diabetes_X = diabetes_X[:, np.newaxis, 2]
diabetes_X = diabetes_X[:, np.newaxis]
print(diabetes_X.shape, diabetes_y.shape)
````

````python
# Linear regression
linear_regression = LinearRegression()

# Train the model
linear_regression.fit( diabetes_X.reshape (-1,1), diabetes_y )
````

````python
# Make predictions using the testing set
diabetes_y_pred = linear_regression.predict( diabetes_X.reshape(-1,1) )


print ("w = " + str( linear_regression.coef_ ) + ", b = " + str( linear_regression.intercept_ ))
````

````python
_, ax = plt.subplots (figsize = ( 5 , 4 ))
ax.scatter( diabetes_X.squeeze() , diabetes_y, color = 'blue') # Squeeze diabetes_X to 1D
ax.plot ( diabetes_X.squeeze() , diabetes_y_pred, color = 'red', linestyle = "--") # Squeeze diabetes_X to 1D
````
