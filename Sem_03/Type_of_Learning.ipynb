## ðŸ”¥ Exercise: Identify the Type of Learning ðŸ¤–

**ðŸŽ¯ Objective:** Write a Python function that classifies a given problem into Supervised, Unsupervised, or Semi-supervised Learning based on the dataset description.

**ðŸ“ Instructions:**

1. Define a function ```identify_learning_type(description: str)```.

2. The function should check for keywords such as:
  - ðŸ“Œ *Supervised Learning* â†’ If the description contains any of these words:
  "labels", "target", "ground truth", "training examples", "annotation", the function returns "Supervised Learning".
  - ðŸ“Œ *Unsupervised Learning* â†’ If it contains:
  "clusters", "patterns", "grouping", "hidden structure", "data distribution", the function returns "Unsupervised Learning".
  - ðŸ“Œ *Semi-supervised Learning* â†’ If it contains:
  "few labels", "unlabeled data", "limited supervision", "pseudo-labels", "partially annotated", the function returns "Semi-supervised Learning".
  
3. If none of these keywords are found, the function returns "Unknown Learning Type".


ðŸš€ Test the function with a different example dataset descriptions.

  ```
  # Example Tests
  print(identify_learning_type("This dataset has input features and corresponding labels."))

  print(identify_learning_type("The goal is to find hidden structure and patterns in the data."))
  ```

````python
def tipo_de_aprendizaje(descripcion: str) -> str:

  descripcion = descripcion.lower()

  if "labels" in descripcion or "target" in descripcion or "ground truth" in descripcion or "training examples" in descripcion or "annotation" in descripcion:
    return "Supervised Learning"
  elif "clusters" in descripcion or "patterns" in descripcion or "grouping" in descripcion or "hidden structure" in descripcion or "data distribution" in descripcion:
    return "Unsupervised Learning"
  elif "few labels" in descripcion or "unlabeled data" in descripcion or "limited supervision" in descripcion or "pseudo-labels" in descripcion or "partially annotated" in descripcion:
    return "Semi-supervised Learning"
  else:
    return "Unknown Learning Type"

text = input("Add a description: ")
print(tipo_de_aprendizaje(text))
````
